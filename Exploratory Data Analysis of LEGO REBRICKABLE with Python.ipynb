# MERGE DATA

Joining the 2 dataset to answer questions

## Lets answer some questions

1) whats the total numbers of parts per theme

```python
Num_Parts_per_Themes=sets_themes.groupby('name_themes').sum().reset_index()
Num_Parts_per_Themes=Num_Parts_per_Themes[['name_themes','num_parts']].sort_values('num_parts' , ascending = False)
Num_Parts_per_Themes
```

2) whats the total number of parts per year

```python
Num_Parts_per_Year=sets_themes.groupby('year').sum().reset_index()
Num_Parts_per_Year=Num_Parts_per_Year[['year','num_parts']].sort_values('num_parts' , ascending = False)
Num_Parts_per_Year
```

3) How many sets were created in each century in the dataset

we have to create century as a derived column to the sets_themes dataset

```python
unique_year=sets_themes['year'].unique()
unique_year.sort()
unique_year
```

```python
sets_themes['year'] > 2000
sets_themes['century']= np.where(sets_themes['year'] > 2000,'21st_century', '20th_century')
sets_themes
```

```python
Num_set_per_century=sets_themes.groupby('century').count().reset_index()
Num_set_per_century[['century','set_num']]
```

9) whats the most produced color of LEGO in term of quantity of parts

Note: Joining two(2) dataset to answer question

```python
colors=pd.read_csv(r"C:\Users\HP\Desktop\DATASET\Rebrickable datasource\LEGO\colors.csv")
colors
```

```python
inventory_parts=pd.read_csv(r"C:\Users\HP\Desktop\DATASET\Rebrickable datasource\LEGO\inventory_parts.csv")
inventory_parts
```

```python
color_inventory=pd.merge(inventory_parts, colors, how = 'left', left_on = 'color_id', right_on = 'id')
color_inventory
```

```python
color_inventory.rename(columns={'name':'color_name'}, inplace = True)
color_inventory
```

```python
popular_color=color_inventory.groupby('color_name').sum().reset_index()
popular_color=popular_color[['color_name','quantity']].sort_values('quantity', ascending = False)
popular_color.head(50)
```
